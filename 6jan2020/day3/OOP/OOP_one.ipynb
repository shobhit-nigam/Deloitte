```python
class biryani:
    def __init__(self, p='dum', r='basmati', s='lots'):
        self.prep = p
        self.rice = r
        self.spice = s
    def make(self):
        print("slow & flavoured")
    def eat(self):
        print("eat it quickly")
    def display(self):
        print("spice is ", self.spice)
        print("prep is ", self.prep)
        print("rice is ", self.rice)
```
