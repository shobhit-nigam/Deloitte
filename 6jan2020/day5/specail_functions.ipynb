```python
def sqr(la):
    return la*la
```

```python
lista = [2, 3, 7, 9]
```

```python
listb = map(sqr, lista)
```

```python
print(list(listb))
```

```python
def funca(la):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if la in vowels:
        return True
    else:
        return False
```

```python
stra = 'honda activa'
listc = list(stra)
```

```python
listc
```

```python
listd = list(filter(funca, listc))
```

```python
listd
```
